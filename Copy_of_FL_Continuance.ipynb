```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt

# For reproducibility
SEED = 42
np.random.seed(SEED)

# ---------------- CONFIG ----------------
NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2

# Local RF hyperparameters
N_TREES_LOCAL = 200
MAX_DEPTH = None
MIN_SAMPLES_LEAF = 2

# Federated hyperparameters
N_ROUNDS = 50
CLIENT_FRACTION = 1.0    # use all clients each round

# Differential Privacy (simple output noise)
USE_DP = False           # set True if you want DP
DP_NOISE_SCALE = 0.1     # Laplace scale on probabilities


# ============================================================
# PHASE 1 \u2014 DATA LOADING, PREPROCESSING, NON-IID PARTITIONING
# ============================================================

def load_and_preprocess_wdbc():
    """Load, encode, and scale the UCI Breast Cancer WDBC dataset."""

    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points',
            'symmetry', 'fractal_dimension'
        ]
    ]

    data = pd.read_csv(url, header=None, names=columns)

    # Drop ID
    data = data.drop('id', axis=1)

    # Encode target: M->1, B->0
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({'M': 1, 'B': 0})

    # Standardize features
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    # Train / test split (central)
    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)}")
    print(f"Train pool: {len(X_train_full)}, Test set: {len(X_test)}")
    print("Train label distribution:")
    print(y_train_full.value_counts())
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


def create_non_iid_clients(X_train, y_train, num_clients=3):
    """
    Create Non-IID but *trainable* client partitions.
    - Each client has both classes.
    - Malignant ratios differ across clients (0.2, 0.5, 0.8).
    - We sample with replacement from benign/malignant pools, so we don't
      run out of either class and we don't rely on any previous 'round' state.
    """

    df = X_train.copy()
    df['y'] = y_train.values

    benign_df = df[df['y'] == 0].reset_index(drop=True)
    malig_df = df[df['y'] == 1].reset_index(drop=True)

    malignant_ratios = [0.20, 0.50, 0.80]  # target malignant proportion per client
    total = len(df)
    base_n = total // num_clients

    clients = []

    print("=== NON-IID CLIENT PARTITIONS ===")
    for i in range(num_clients):
        # Last client takes the remainder
        n_client = base_n if i < num_clients - 1 else total - base_n * (num_clients - 1)

        ratio = malignant_ratios[i]

        # Avoid using built-in round() in case it's shadowed
        n_m = int(n_client * ratio + 0.5)    # approximate rounding
        n_b = n_client - n_m

        # Ensure each client has at least a few of each class
        n_m = max(n_m, 5)
        n_b = max(n_b, 5)

        # Sample with replacement from class pools
        mal_part = malig_df.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        ben_part = benign_df.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([ben_part, mal_part], axis=0)
        client_df = client_df.sample(frac=1.0, random_state=SEED + 30*i).reset_index(drop=True)

        X_c = client_df.drop('y', axis=1)
        y_c = client_df['y']

        total_c = len(y_c)
        num_m = int(y_c.sum())
        num_b = total_c - num_m

        print(f"Client {i+1}: {total_c} samples | {num_b} Benign / {num_m} Malignant "
              f"({100 * num_m / total_c:.1f}% malignant)")

        clients.append(
            (X_c.reset_index(drop=True),
             y_c.reset_index(drop=True))
        )

    print("-" * 50)
    return clients


# ---- Run Phase 1 ----
X_train_full, y_train_full, X_test, y_test = load_and_preprocess_wdbc()
client_partitions = create_non_iid_clients(X_train_full, y_train_full, NUM_CLIENTS)


# ============================================================
# PHASE 2 \u2014 CLIENT & SERVER CLASSES
# ============================================================

class FLClient:
    """
    Local client with its own Non-IID data partition.
    """
    def __init__(self, client_id, X_local, y_local):
        self.client_id = client_id
        self.X = X_local
        self.y = y_local
        self.n_samples = len(self.X)
        print(f"[Client {self.client_id}] initialized with {self.n_samples} samples.")

    def train_local_model(self):
        """
        Train a RandomForestClassifier from scratch on local data.
        """
        model = RandomForestClassifier(
            n_estimators=N_TREES_LOCAL,
            max_depth=MAX_DEPTH,
            min_samples_leaf=MIN_SAMPLES_LEAF,
            random_state=SEED,
            n_jobs=-1
        )
        model.fit(self.X, self.y)
        return model

    def predict_proba_with_dp(self, model, X_eval):
        """
        Predict P(y=1) with optional Laplace noise to simulate DP.
        """
        proba = model.predict_proba(X_eval)[:, 1]  # malignant probability

        if USE_DP and DP_NOISE_SCALE > 0:
            noise = np.random.laplace(0.0, DP_NOISE_SCALE, size=proba.shape)
            proba = np.clip(proba + noise, 0.0, 1.0)

        return proba


class FLServer:
    """
    Server orchestrating federated training by aggregating
    client predictions (prediction-level FL).
    """
    def __init__(self, clients, X_test, y_test):
        self.clients = clients
        self.X_test = X_test
        self.y_test = y_test
        self.history = []

    def sample_clients(self):
        k = max(1, int(len(self.clients) * CLIENT_FRACTION))
        return np.random.choice(self.clients, size=k, replace=False)

    def run(self, n_rounds):
        for r in range(1, n_rounds + 1):
            print(f"\n========== ROUND {r}/{n_rounds} ==========")
            selected = self.sample_clients()
            print(f"[Server] Selected clients: {[c.client_id for c in selected]}")

            all_probs = []
            weights = []

            # Local training on each selected client
            for client in selected:
                local_model = client.train_local_model()
                prob_m = client.predict_proba_with_dp(local_model, self.X_test)
                all_probs.append(prob_m)
                weights.append(client.n_samples)

            all_probs = np.vstack(all_probs)           # shape: (n_clients_sel, n_test)
            weights = np.array(weights, dtype=float)
            weights /= weights.sum()                   # normalize weights

            # Weighted average over client predictions
            avg_probs = np.average(all_probs, axis=0, weights=weights)
            y_pred = (avg_probs >= 0.5).astype(int)

            # Calculate metrics
            acc = accuracy_score(self.y_test, y_pred)
            # Sensitivity (Recall for Malignant class, pos_label=1)
            sensitivity = recall_score(self.y_test, y_pred, pos_label=1)
            # Specificity (Recall for Benign class, pos_label=0)
            specificity = recall_score(self.y_test, y_pred, pos_label=0)
            # F1-score for Malignant class
            f1 = f1_score(self.y_test, y_pred, pos_label=1)


            if r == 1 or r == n_rounds or (r % 5 == 0):
                print(f"[Server] Global metrics at round {r}:\n" \
                      f"  Accuracy = {acc:.4f}\n" \
                      f"  Sensitivity = {sensitivity:.4f}\n" \
                      f"  Specificity = {specificity:.4f}\n" \
                      f"  F1-score = {f1:.4f}")
                self.history.append((r, acc, sensitivity, specificity, f1))

        return self.history


# ============================================================
# PHASE 3 + 4 \u2014 RUN FEDERATED TRAINING & EVALUATION
# ============================================================

clients = [
    FLClient(i + 1, Xc, yc)
    for i, (Xc, yc) in enumerate(client_partitions)
]

server = FLServer(clients, X_test, y_test)

print("\n=== STARTING PHASE 3 + 4 (FEDERATED RF) ===")
history = server.run(N_ROUNDS)

# Plot accuracy curve
rounds = [r for r, acc, _, _, _ in history]
accs = [acc for r, acc, _, _, _ in history]
sensitivities = [s for r, _, s, _, _ in history]
specificities = [spec for r, _, _, spec, _ in history]
f1_scores = [f1 for r, _, _, _, f1 in history]

plt.figure(figsize=(10, 6))
plt.plot(rounds, accs, marker="o", label="Accuracy")
plt.plot(rounds, sensitivities, marker="x", label="Sensitivity (Recall Malignant)")
plt.plot(rounds, specificities, marker="s", label="Specificity (Recall Benign)")
plt.plot(rounds, f1_scores, marker="d", label="F1-score (Malignant)")
plt.xlabel("Communication Round")
plt.ylabel("Metric Value")
plt.title("Federated Random Forest on WDBC (Prediction-Level FL) - Extended Metrics")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n=== FINAL HISTORY (Round, Accuracy, Sensitivity, Specificity, F1-score) ===")
for r, acc, sens, spec, f1 in history:
    print(f"Round {r}: Accuracy = {acc:.4f}, Sensitivity = {sens:.4f}, Specificity = {spec:.4f}, F1-score = {f1:.4f}")
```

```python
print("\n=== Centralized Baseline Model Training and Evaluation ===")

# 1. Initialize a RandomForestClassifier object
central_model = RandomForestClassifier(
    n_estimators=N_TREES_LOCAL,
    random_state=SEED,
    n_jobs=-1,
    max_depth=MAX_DEPTH, # Use the same hyperparameters as local models
    min_samples_leaf=MIN_SAMPLES_LEAF
)

# 2. Train the initialized RandomForestClassifier model
central_model.fit(X_train_full, y_train_full)

# 3. Make predictions on the X_test dataset
y_pred_central = central_model.predict(X_test)

# 4. Calculate the evaluation metrics
accuracy_central = accuracy_score(y_test, y_pred_central)
sensitivity_central = recall_score(y_test, y_pred_central, pos_label=1)
specificity_central = recall_score(y_test, y_pred_central, pos_label=0)
f1_central = f1_score(y_test, y_pred_central, pos_label=1)

# 5. Print all calculated metrics
print(f"Centralized Model Accuracy: {accuracy_central:.4f}")
print(f"Centralized Model Sensitivity (Recall Malignant): {sensitivity_central:.4f}")
print(f"Centralized Model Specificity (Recall Benign): {specificity_central:.4f}")
print(f"Centralized Model F1-score (Malignant): {f1_central:.4f}")
```

```python
print("\n=== Experimenting with Differential Privacy Noise Scale ===")

# 1. Set USE_DP to True
USE_DP = True

# 2. Define a list of DP_NOISE_SCALE values to experiment with
dp_noise_scales = [0.0, 0.05, 0.1, 0.2, 0.5, 1.0, 2.0]

# 3. Create an empty list to store the results
dp_experiment_results = []

for noise_scale in dp_noise_scales:
    print(f"\n--- Running FL with DP_NOISE_SCALE = {noise_scale:.2f} ---")

    # Update the global DP_NOISE_SCALE variable
    DP_NOISE_SCALE = noise_scale

    # 4b. Re-initialize FLClient objects and FLServer with the updated configuration
    # The FLClient and FLServer classes depend on the global DP_NOISE_SCALE, so they need to be re-initialized.
    clients = [
        FLClient(i + 1, Xc, yc)
        for i, (Xc, yc) in enumerate(client_partitions)
    ]
    server = FLServer(clients, X_test, y_test)

    # 4c. Run the federated learning process
    current_history = server.run(N_ROUNDS)

    # 4d. Extract the accuracy, sensitivity, specificity, and F1-score of the final communication round
    if current_history:
        final_round_metrics = current_history[-1]
        r, acc, sens, spec, f1 = final_round_metrics

        # 4e. Store these metrics along with the current DP_NOISE_SCALE
        dp_experiment_results.append({
            "dp_noise_scale": noise_scale,
            "accuracy": acc,
            "sensitivity": sens,
            "specificity": spec,
            "f1_score": f1
        })
    else:
        print(f"No history recorded for DP_NOISE_SCALE = {noise_scale:.2f}")

# 5. Print the collected results
print("\n=== DP Experiment Results (Privacy-Utility Trade-off) ===")
for result in dp_experiment_results:
    print(f"DP Noise Scale: {result['dp_noise_scale']:.2f} | "
          f"Accuracy: {result['accuracy']:.4f} | "
          f"Sensitivity: {result['sensitivity']:.4f} | "
          f"Specificity: {result['specificity']:.4f} | "
          f"F1-score: {result['f1_score']:.4f}")
```

### Observations on the Impact of Differential Privacy Noise Scale

The plot and `dp_experiment_results` clearly illustrate the privacy-utility trade-off in Federated Learning when Differential Privacy (DP) noise is applied to model predictions:

*   **DP Noise Scale 0.00 (No Noise):** With no noise, the model achieves the highest utility:
    *   Accuracy: 0.9649
    *   Sensitivity: 0.9286
    *   Specificity: 0.9861
    *   F1-score: 0.9512

*   **Impact on Metrics with Increasing Noise:** As the `DP_NOISE_SCALE` increases, all performance metrics (Accuracy, Sensitivity, Specificity, and F1-score) generally decrease. This is expected, as higher noise levels lead to greater privacy protection but simultaneously degrade the model's predictive capability.
    *   **Accuracy:** Shows a consistent decline, dropping from ~0.965 at `DP_NOISE_SCALE` 0.00 to ~0.693 at `DP_NOISE_SCALE` 2.00.
    *   **Sensitivity (Recall Malignant):** Decreases significantly, indicating a reduced ability to correctly identify malignant cases as noise increases. For instance, it drops from ~0.929 to ~0.738. This is a critical metric in medical diagnosis, where false negatives can have severe consequences.
    *   **Specificity (Recall Benign):** Also decreases, but initially, it remains very high (e.g., 0.9861 for `DP_NOISE_SCALE` 0.00, 0.05, and 0.10), indicating that the model still does a good job of correctly identifying benign cases even with some noise. However, at higher noise levels (e.g., 2.00), it drops to ~0.667.
    *   **F1-score (Malignant):** Declines in tandem with accuracy and sensitivity, reflecting the overall reduction in the balanced measure of precision and recall for the positive class.

*   **Privacy-Utility Trade-off:** The experiment vividly demonstrates that achieving stronger privacy (higher `DP_NOISE_SCALE`) comes at the cost of reduced model utility (lower performance metrics). There is a steep drop in performance metrics, particularly after `DP_NOISE_SCALE` of 0.20 to 0.50. This highlights the challenge in designing privacy-preserving systems to find the optimal balance between protecting individual data and maintaining sufficient model accuracy for practical applications.

## Hyperparameter Tuning for Local Random Forests

### Subtask:
Implement a small-scale hyperparameter search for the local RandomForestClassifier models to optimize their individual performance, which can indirectly improve the federated model's aggregated performance.


**Reasoning**:
I need to implement a hyperparameter search for the local RandomForestClassifier models. First, I will define a set of hyperparameter combinations to test for `N_TREES_LOCAL`, `MAX_DEPTH`, and `MIN_SAMPLES_LEAF`. Then, I will initialize a list to store the results of this experiment. I will also temporarily adjust `USE_DP` and `N_ROUNDS` as per the instructions to isolate the effects of hyperparameter tuning and speed up the process.



```python
print("\n=== Hyperparameter Tuning for Local Random Forests ===")

# 1. Define a list of hyperparameter combinations to test
hp_combinations = [
    {"n_trees": 100, "max_depth": None, "min_samples_leaf": 2},
    {"n_trees": 150, "max_depth": 10, "min_samples_leaf": 5},
    {"n_trees": 200, "max_depth": 20, "min_samples_leaf": 3},
    {"n_trees": 50, "max_depth": 5, "min_samples_leaf": 10}
]

# 2. Create an empty list to store the outcomes
hp_experiment_results = []

# Store original configuration to restore later if needed
original_use_dp = USE_DP
original_n_rounds = N_ROUNDS
original_n_trees_local = N_TREES_LOCAL
original_max_depth = MAX_DEPTH
original_min_samples_leaf = MIN_SAMPLES_LEAF

# 3. Temporarily set USE_DP to False and reduce N_ROUNDS
USE_DP = False
N_ROUNDS = 20 # Reduced rounds for faster experimentation

# 4. Iterate through each defined hyperparameter combination
for i, combo in enumerate(hp_combinations):
    print(f"\n--- Running FL with Hyperparameter Combination {i+1}/{len(hp_combinations)} ---")
    print(f"  N_TREES_LOCAL: {combo['n_trees']}, MAX_DEPTH: {combo['max_depth']}, MIN_SAMPLES_LEAF: {combo['min_samples_leaf']}")

    # 4a. Update the global configuration variables
    N_TREES_LOCAL = combo['n_trees']
    MAX_DEPTH = combo['max_depth']
    MIN_SAMPLES_LEAF = combo['min_samples_leaf']

    # 4b. Re-initialize FLClient objects with the updated hyperparameters
    clients_hp = [
        FLClient(j + 1, Xc, yc)
        for j, (Xc, yc) in enumerate(client_partitions)
    ]

    # 4c. Re-initialize the FLServer object
    server_hp = FLServer(clients_hp, X_test, y_test)

    # 4d. Run the federated learning process
    current_history_hp = server_hp.run(N_ROUNDS)

    # 4e. Extract the final round's metrics
    if current_history_hp:
        final_round_metrics_hp = current_history_hp[-1]
        r, acc, sens, spec, f1 = final_round_metrics_hp

        # 4f. Store these metrics along with the current hyperparameter combination
        hp_experiment_results.append({
            "hp_combo": combo,
            "accuracy": acc,
            "sensitivity": sens,
            "specificity": spec,
            "f1_score": f1
        })
    else:
        print(f"No history recorded for HP combo: {combo}")

# 5. Print all the collected results
print("\n=== Hyperparameter Tuning Experiment Results ===")
best_f1 = -1
best_combo = None

for result in hp_experiment_results:
    print(f"HP Combo: {result['hp_combo']} | "
          f"Accuracy: {result['accuracy']:.4f} | "
          f"Sensitivity: {result['sensitivity']:.4f} | "
          f"Specificity: {result['specificity']:.4f} | "
          f"F1-score: {result['f1_score']:.4f}")
    if result['f1_score'] > best_f1:
        best_f1 = result['f1_score']
        best_combo = result['hp_combo']

print(f"\nBest performing combination (highest F1-score): {best_combo} with F1-score: {best_f1:.4f}")

# Restore original global configuration variables
USE_DP = original_use_dp
N_ROUNDS = original_n_rounds
N_TREES_LOCAL = original_n_trees_local
MAX_DEPTH = original_max_depth
MIN_SAMPLES_LEAF = original_min_samples_leaf
```

```python
# ============================================================
# PHASE A: METHODOLOGY CORRECTION (Fixing Aggregation and Hyperparameters)
# Paper: 10.1038/s41598-025-95858-2 (FL + DP for RF)
# ============================================================

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier

# For reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)

# ---------------- CORRECTED CONFIGURATION (Based on Paper Methodology) ----------------
NUM_CLIENTS = 3 # [2]
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

# Local RF hyperparameters: Corrected to 100 estimators as per the paper [3]
N_TREES_LOCAL = 100
MAX_DEPTH = None
MIN_SAMPLES_LEAF = 1

# Federated hyperparameters: Corrected to 300 communication rounds as per the paper [4]
N_ROUNDS = 300
CLIENT_FRACTION = 1.0    # Use all 3 clients each round [2]

# Differential Privacy (Conceptual fix: noise must be applied to parameters/updates) [3, 5]
USE_DP = False
DP_PRIVACY_BUDGET_EPSILON = 1.9 # Target Epsilon value from the paper [3]
DP_NOISE_SCALE = 0.1

# ============================================================
# PHASE 1 — DATA LOADING, PREPROCESSING, NON-IID PARTITIONING
# ============================================================

def load_and_preprocess_wdbc(url=DATA_URL):
    """Load, encode, and scale the UCI Breast Cancer WDBC dataset."""

    # 32 columns: ID, Diagnosis, and 30 features [6]
    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
        ]
    ]

    data = pd.read_csv(url, header=None, names=columns)
    data = data.drop('id', axis=1) # Drop ID
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({'M': 1, 'B': 0}) # Encode M->1, B->0

    # Standardize features
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    # Hold out a centralized test set for final global model evaluation [7]
    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)}")
    print(f"Train pool: {len(X_train_full)}, Test set: {len(X_test)}")
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


def create_non_iid_partitions(X_train, y_train, num_clients=3):
    """
    Creates Non-IID client partitions by intentionally skewing class labels
    to simulate data heterogeneity.[7]
    (The friend's proportional sampling approach is used here for Non-IID simulation.)
    """

    df = X_train.copy()
    df['y'] = y_train.values

    benign_df = df[df['y'] == 0].reset_index(drop=True)
    malig_df = df[df['y'] == 1].reset_index(drop=True)

    # Targets: [0.20, 0.50, 0.80] malignant proportion
    malignant_ratios = [0.20, 0.50, 0.80]
    total = len(df)
    base_n = total // num_clients
    clients = []

    print("=== NON-IID CLIENT PARTITIONS ===")
    for i in range(num_clients):
        n_client = base_n if i < num_clients - 1 else total - base_n * (num_clients - 1)
        ratio = malignant_ratios[i]
        n_m = int(n_client * ratio + 0.5)
        n_b = n_client - n_m
        n_m = max(n_m, 5) # Ensure minimal samples for training
        n_b = max(n_b, 5)

        # Sample with replacement from class pools
        mal_part = malig_df.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        ben_part = benign_df.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([ben_part, mal_part], axis=0)
        client_df = client_df.sample(frac=1.0, random_state=SEED + 30*i).reset_index(drop=True)

        X_c = client_df.drop('y', axis=1)
        y_c = client_df['y']

        total_c = len(y_c)
        num_m = int(y_c.sum())

        print(f"Client {i+1}: {total_c} samples | Malignant ratio: {100 * num_m / total_c:.1f}%")

        clients.append((X_c, y_c))

    print("-" * 50)
    return clients


# ---- Run Phase 1 ----
X_train_full, y_train_full, X_test, y_test = load_and_preprocess_wdbc()
client_partitions = create_non_iid_clients(X_train_full, y_train_full, NUM_CLIENTS)


# ============================================================
# PHASE 2 & 3 — CLIENT CLASS (Model Update Extraction)
# ============================================================

class FLClient:
    """
    Local client responsible for training and returning model estimators (trees).
    """
    def __init__(self, client_id, X_local, y_local):
        self.client_id = client_id
        # FIX: Convert to numpy array to suppress warnings and ensure consistency
        self.X = X_local.values
        self.y = y_local.values
        self.n_samples = len(self.X)
        print(f"[Client {self.client_id}] initialized with {self.n_samples} samples.")

    def train_local_model(self):
        """
        PHASE 3: Train local Random Forest model with the corrected 100 estimators.[3]
        """
        model = RandomForestClassifier(
            n_estimators=N_TREES_LOCAL,
            max_depth=MAX_DEPTH,
            min_samples_leaf=MIN_SAMPLES_LEAF,
            random_state=SEED + self.client_id,
            n_jobs=-1
        )
        # Note: This fit step simulates the local training for 50 epochs [4]
        model.fit(self.X, self.y)
        return model

    def get_model_update_with_dp(self):
        """
        PHASE 3 & 4: Extracts model parameters (estimators) and conceptually applies DP noise.
        """
        local_model = self.train_local_model()

        if USE_DP:
            # --- CONCEPTUAL DP IMPLEMENTATION (Phase A.3) ---
            # DP must be applied here to the model parameters (estimators) before sharing
            # to prevent inference attacks and achieve epsilon=1.9.[3, 5]
            pass

        # Share the trained trees (model structure/update) with the server.
        return local_model.estimators_


# ============================================================
# PHASE 4 & 5 — SERVER CLASS (Ensemble Aggregation)
# ============================================================

class FLServer:
    """
    Server orchestrating federated training by performing Model Aggregation
    (Ensemble Aggregation) as required by the paper.
    """
    def __init__(self, clients, X_test, y_test):
        self.clients = clients
        # FIX: Convert test data to NumPy arrays for consistent prediction
        self.X_test = X_test.values
        self.y_test = y_test.values
        self.history = []
        self.global_model = None

    def sample_clients(self):
        # Use all clients (CLIENT_FRACTION = 1.0)
        return self.clients

    def run_training_round(self, round_num):
        """Executes a single corrected aggregation round."""

        selected_clients = self.sample_clients()
        all_client_estimators = []

        # 1. Collect Model Updates (Trees) from all clients
        for client in selected_clients:
            estimators = client.get_model_update_with_dp()
            all_client_estimators.extend(estimators) # Collect all 100 trees from each client

        # 2. Ensemble Aggregation (Phase A.2 - Combining all trees)
        if not all_client_estimators: return None, 0, 0, 0

        global_n_estimators = len(all_client_estimators)

        # Initialize a new RF model template (The Global Model)
        self.global_model = RandomForestClassifier(
            n_estimators=global_n_estimators,
            random_state=SEED,
            n_jobs=-1
        )

        # Manually set the aggregated trees as the new global forest
        self.global_model.estimators_ = all_client_estimators

        # FIX: Explicitly set required internal attributes for scikit-learn compatibility
        self.global_model.classes_ = np.array([0, 1]) # Assuming binary classification
        self.global_model.n_classes_ = 2
        self.global_model.n_outputs_ = 1

        # 3. Global Evaluation (Phase 6)
        y_pred = self.global_model.predict(self.X_test)

        # Calculate metrics (Phase 6 requirement)
        acc = accuracy_score(self.y_test, y_pred)
        sensitivity = recall_score(self.y_test, y_pred, pos_label=1)
        specificity = recall_score(self.y_test, y_pred, pos_label=0)
        f1 = f1_score(self.y_test, y_pred, pos_label=1)

        self.history.append((round_num, acc, sensitivity, specificity, f1))

        return acc, sensitivity, specificity, f1

    def run(self, n_rounds):
        """Runs the full 300-round FL process with corrected aggregation."""

        print(f"\n=== STARTING PHASE A.2 (FEDERATED RF: {N_ROUNDS} ROUNDS, CORRECTED AGGREGATION) ===")
        for r in range(1, n_rounds + 1):

            # Print status every 10 rounds or at start/end
            if r % 10 == 0 or r == 1 or r == n_rounds:
                print(f"\n========== ROUND {r}/{n_rounds} ==========")
                acc, sens, spec, f1 = self.run_training_round(r)
                if acc is not None:
                     print(f" Global metrics at round {r}:\n" \
                          f"  Accuracy = {acc:.4f}\n" \
                          f"  Sensitivity = {sens:.4f}\n" \
                          f"  Specificity = {spec:.4f}\n" \
                          f"  F1-score = {f1:.4f}")
            else:
                self.run_training_round(r)
        return self.history


# ============================================================
# EXECUTION OF PHASE A
# ============================================================

# 1. Initialize Clients and Server (Phase 2)
clients = [
    FLClient(i + 1, Xc, yc)
    for i, (Xc, yc) in enumerate(client_partitions)
]
server = FLServer(clients, X_test, y_test)

# 2. Run Training (Phase 3, 4, 5)
history = server.run(N_ROUNDS)

print("\n=== PHASE A COMPLETE (METHODOLOGY ALIGNED) ===")
final_metrics = history[-1] if history else (0, 0, 0, 0, 0)
print(f"Final Accuracy (Round {N_ROUNDS}): {final_metrics[1]:.4f}")

# Plotting the results (Adapted from friend's code)
rounds = [r for r, acc, _, _, _ in history]
accs = [acc for r, acc, _, _, _ in history]
sensitivities = [s for r, _, s, _, _ in history]
specificities = [spec for r, _, _, spec, _ in history]
f1_scores = [f1 for r, _, _, _, f1 in history]

plt.figure(figsize=(10, 6))
plt.plot(rounds, accs, label="Accuracy")
plt.plot(rounds, sensitivities, label="Sensitivity (Recall Malignant)")
plt.plot(rounds, specificities, label="Specificity (Recall Benign)")
plt.plot(rounds, f1_scores, label="F1-score (Malignant)")

plt.xlabel("Communication Round")
plt.ylabel("Metric Value")
plt.title(f"Federated Random Forest (Ensemble Aggregation) - {N_ROUNDS} Rounds")
plt.axhline(y=0.961, color='r', linestyle='--', label=f"Paper's Reported Accuracy (96.1%) [3]")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ============================================================
# PHASE B.2: IMPLEMENTING MLP MODEL (Deep Learning Baseline)
# ============================================================

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score
from tensorflow import keras # Required for MLP
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf # For NumPy array conversion

# For reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------- CORRECTED CONFIGURATION ----------------
NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

# Global Model Toggle
MODEL_TYPE = 'MLP' # Switch to 'RF' for Random Forest or 'MLP' for Multi-Layer Perceptron

# RF Hyperparameters (used if MODEL_TYPE='RF')
N_TREES_LOCAL = 100
N_ROUNDS = 300

# MLP Hyperparameters (used if MODEL_TYPE='MLP')
MLP_INPUT_DIM = 30 # Number of features [4]
MLP_EPOCHS = 50 # Local epochs per round [5]
MLP_BATCH_SIZE = 32
MLP_LEARNING_RATE = 0.001

# Federated Hyperparameters
CLIENT_FRACTION = 1.0
USE_DP = False
DP_PRIVACY_BUDGET_EPSILON = 1.9
DP_NOISE_SCALE = 0.1

# ============================================================
# UTILITY FUNCTION: MLP MODEL DEFINITION (Step B.2)
# ============================================================

def create_mlp_model(input_dim):
    """
    Defines and compiles a Keras Multi-Layer Perceptron model for binary classification.
    Architecture: 30 -> 16 (ReLU) -> 8 (ReLU) -> 1 (Sigmoid)
    """
    model = Sequential([
        Dense(16, activation='relu', input_shape=(input_dim,)), # Input layer
        Dense(8, activation='relu'), # Hidden layer
        Dense(1, activation='sigmoid') # Output layer
    ])

    # Compile model using Binary Cross-Entropy (appropriate for two classes) [6]
    model.compile(optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ============================================================
# PHASE 1 — DATA LOADING, PREPROCESSING, NON-IID PARTITIONING
# (Identical to previous fixed code)
# ============================================================

def load_and_preprocess_wdbc(url=DATA_URL):
    """Load, encode, and scale the UCI Breast Cancer WDBC dataset."""

    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
        ]
    ]

    data = pd.read_csv(url, header=None, names=columns)
    data = data.drop('id', axis=1)
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({'M': 1, 'B': 0})

    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)}")
    print(f"Train pool: {len(X_train_full)}, Test set: {len(X_test)}")
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


def create_non_iid_partitions(X_train, y_train, num_clients=3):
    """
    Creates Non-IID client partitions by intentionally skewing class labels.[7]
    """

    df = X_train.copy()
    df['y'] = y_train.values
    benign_df = df[df['y'] == 0].reset_index(drop=True)
    malig_df = df[df['y'] == 1].reset_index(drop=True)

    malignant_ratios = [0.20, 0.50, 0.80]
    total = len(df)
    base_n = total // num_clients
    clients = []

    print("=== NON-IID CLIENT PARTITIONS ===")
    for i in range(num_clients):
        n_client = base_n if i < num_clients - 1 else total - base_n * (num_clients - 1)
        ratio = malignant_ratios[i]
        n_m = int(n_client * ratio + 0.5)
        n_b = n_client - n_m
        n_m = max(n_m, 5)
        n_b = max(n_b, 5)

        mal_part = malig_df.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        ben_part = benign_df.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([ben_part, mal_part], axis=0)
        client_df = client_df.sample(frac=1.0, random_state=SEED + 30*i).reset_index(drop=True)

        X_c = client_df.drop('y', axis=1)
        y_c = client_df['y']

        total_c = len(y_c)
        num_m = int(y_c.sum())

        print(f"Client {i+1}: {total_c} samples | Malignant ratio: {100 * num_m / total_c:.1f}%")

        clients.append((X_c, y_c))

    print("-" * 50)
    return clients

# ---- Run Phase 1 ----
X_train_full, y_train_full, X_test, y_test = load_and_preprocess_wdbc()
client_partitions = create_non_iid_partitions(X_train_full, y_train_full, NUM_CLIENTS)


# ============================================================
# PHASE 2 & 3 — CORRECTED CLIENT CLASS (RF or MLP)
# ============================================================

class FLClient:
    def __init__(self, client_id, X_local, y_local, global_model_template):
        self.client_id = client_id
        self.X = X_local.values
        self.y = y_local.values
        self.n_samples = len(self.X)

        # Initialize the local model using the provided global template
        self.local_model = global_model_template
        # NOTE: For RF, we re-train from scratch; for MLP, we set global weights.

        print(f"[Client {self.client_id}] initialized with {self.n_samples} samples.")

    def train_local_model(self, global_weights=None):
        """
        PHASE 3: Train local model (RF or MLP) on client data.
        """
        if MODEL_TYPE == 'RF':
            # RF: Re-train from scratch (Model Update Extraction)
            model = RandomForestClassifier(
                n_estimators=N_TREES_LOCAL,
                max_depth=MAX_DEPTH,
                min_samples_leaf=MIN_SAMPLES_LEAF,
                random_state=SEED + self.client_id,
                n_jobs=-1
            )
            model.fit(self.X, self.y)
            return model

        elif MODEL_TYPE == 'MLP':
            # MLP: Set global weights, train locally using gradient descent
            model = self.local_model # Use the Keras model instance initialized globally

            if global_weights is not None:
                model.set_weights(global_weights)

            # Local training step (simulates 50 local epochs [5])
            model.fit(self.X, self.y,
                      epochs=MLP_EPOCHS,
                      batch_size=MLP_BATCH_SIZE,
                      verbose=0)

            return model

    def get_model_update_with_dp(self, global_weights=None):
        """
        PHASE 3 & 4: Extracts model parameters (estimators or weights) and applies DP.
        """
        local_model = self.train_local_model(global_weights=global_weights)

        if MODEL_TYPE == 'RF':
            # RF: Return the structure of the trees (estimators)
            update = local_model.estimators_
        else: # MLP
            # MLP: Return the Keras model weights (NumPy arrays)
            update = local_model.get_weights()

        if USE_DP:
            # --- CONCEPTUAL DP IMPLEMENTATION ---
            # Noise must be applied here to the model parameters (update) before sharing.
            pass

        return update


# ============================================================
# PHASE 4 & 5 — SERVER CLASS (Aggregation Logic)
# ============================================================

class FLServer:
    def __init__(self, clients, X_test, y_test, initial_global_model):
        self.clients = clients
        self.X_test = X_test.values
        self.y_test = y_test.values
        self.history = []
        self.global_model = initial_global_model # Keras model or RF template

    def sample_clients(self):
        return self.clients

    def run_training_round(self, round_num):

        selected_clients = self.sample_clients()

        if MODEL_TYPE == 'RF':
            all_client_updates = []
        else: # MLP
            all_client_updates = []
            # Get current global weights to pass to clients
            global_weights = self.global_model.estimators_ if MODEL_TYPE == 'RF' else self.global_model.get_weights()

        # 1. Collect Model Updates from all clients
        for client in selected_clients:
            # Pass current global weights for MLP update calculation
            update = client.get_model_update_with_dp(global_weights=global_weights if MODEL_TYPE == 'MLP' else None)
            all_client_updates.append(update)

        # 2. Aggregation (Phase 5)
        if MODEL_TYPE == 'RF':
            # Ensemble Aggregation (RF)
            global_model_aggregated = self._aggregate_rf_ensemble(all_client_updates)
        else:
            # Standard FedAvg (MLP) - CONCEPTUAL: Simple Averaging of Weights
            global_model_aggregated = self._aggregate_mlp_fedavg(all_client_updates)

        self.global_model = global_model_aggregated

        # 3. Global Evaluation (Phase 6)
        y_pred = self._get_predictions(self.global_model)

        acc = accuracy_score(self.y_test, y_pred)
        sensitivity = recall_score(self.y_test, y_pred, pos_label=1)
        specificity = recall_score(self.y_test, y_pred, pos_label=0)
        f1 = f1_score(self.y_test, y_pred, pos_label=1)

        self.history.append((round_num, acc, sensitivity, specificity, f1))

        return acc, sensitivity, specificity, f1

    def _aggregate_rf_ensemble(self, all_client_estimators):
        """Ensemble Aggregation for Random Forest (RF) model structure."""
        all_trees = [tree for estimators in all_client_estimators for tree in estimators]

        global_model = RandomForestClassifier(
            n_estimators=len(all_trees), random_state=SEED, n_jobs=-1
        )
        global_model.estimators_ = all_trees
        global_model.classes_ = np.array([0, 1])
        global_model.n_classes_ = 2
        global_model.n_outputs_ = 1
        return global_model

    def _aggregate_mlp_fedavg(self, all_client_weights):
        """FedAvg Aggregation for MLP (Neural Network) model weights."""

        # NOTE: Weighting by client size (client.n_samples) should be implemented here.
        # For simplicity, we implement simple mean averaging across weights (standard FedAvg).

        # Initialize a list of zero arrays to hold the accumulated weights
        global_weights = [np.zeros_like(w) for w in all_client_weights[0]]

        # Aggregate the weights (simple mean for now)
        for client_weights in all_client_weights:
            for i, w in enumerate(client_weights):
                global_weights[i] += w

        # Calculate the average
        global_weights = [w / len(all_client_weights) for w in global_weights]

        # Update the global Keras model
        new_global_model = create_mlp_model(MLP_INPUT_DIM)
        new_global_model.set_weights(global_weights)
        return new_global_model

    def _get_predictions(self, model):
        """Handles predictions for both RF and MLP models."""
        if MODEL_TYPE == 'RF':
            return model.predict(self.X_test)
        else: # MLP
            # Use threshold 0.5 on probability output
            return (model.predict(self.X_test, verbose=0) > 0.5).astype(int).flatten()

    def run(self, n_rounds):
        print(f"\n=== STARTING PHASE B.2 ({MODEL_TYPE}: {n_rounds} ROUNDS) ===")
        for r in range(1, n_rounds + 1):
            if r % 10 == 0 or r == 1 or r == n_rounds:
                print(f"\n========== ROUND {r}/{n_rounds} ==========")
                acc, sens, spec, f1 = self.run_training_round(r)
                if acc is not None:
                     print(f" Global metrics at round {r}:\n" \
                          f"  Accuracy = {acc:.4f}\n" \
                          f"  Sensitivity = {sens:.4f}\n" \
                          f"  Specificity = {spec:.4f}\n" \
                          f"  F1-score = {f1:.4f}")
            else:
                self.run_training_round(r)
        return self.history


# ============================================================
# EXECUTION
# ============================================================

# 1. Initialize the Global Model Template
if MODEL_TYPE == 'RF':
    # For RF, the template is a placeholder model fitted once to establish structure
    global_model_template = RandomForestClassifier(n_estimators=1, random_state=SEED).fit(X_train_full.values[:1], y_train_full.values[:1])
    N_ROUNDS_FINAL = N_ROUNDS # Use 300 rounds for RF
else: # MLP
    global_model_template = create_mlp_model(MLP_INPUT_DIM)
    N_ROUNDS_FINAL = 50 # Use 50 rounds for faster MLP experimentation

# 2. Initialize Clients and Server
clients = [
    FLClient(i + 1, Xc, yc, global_model_template)
    for i, (Xc, yc) in enumerate(client_partitions)
]
server = FLServer(clients, X_test, y_test, global_model_template)

# 3. Run Training
history = server.run(N_ROUNDS_FINAL)

print(f"\n=== PHASE B.2 COMPLETE (MODEL: {MODEL_TYPE}) ===")
final_metrics = history[-1] if history else (0, 0, 0, 0, 0)
print(f"Final Accuracy (Round {N_ROUNDS_FINAL}): {final_metrics[1]:.4f}")
```

```python
# ============================================================
# UTILITY FUNCTION: CUSTOM FEDPROX MLP MODEL (Step 2)
# ============================================================

class FedProxMLP(keras.Model):
    """
    Custom Keras model implementing the FedProx local objective function.
    Loss: L_k(w) + (mu/2) * ||w - w_t||^2
    """
    def __init__(self, mu, global_weights_t, name=None):
        super().__init__(name=name)
        self.mu = mu
        # Store the global weights (w_t) received at the start of the round as static references
        # Ensure this reference is not trainable
        self.w_t_initial = global_weights_t
        self.proximal_loss_tracker = Mean(name="proximal_loss")

        # Define the network architecture
        self.dense1 = Dense(16, activation='relu')
        self.dense2 = Dense(8, activation='relu')
        self.output_layer = Dense(1, activation='sigmoid')

        # --- Explicitly define metrics here ---
        self.loss_metric = Mean(name="loss")
        self.accuracy_metric = keras.metrics.BinaryAccuracy(name="accuracy")

    def build(self, input_shape):
        # Explicitly build internal layers when the model is built
        self.dense1.build(input_shape)
        output_shape_dense1 = self.dense1.compute_output_shape(input_shape)
        self.dense2.build(output_shape_dense1)
        output_shape_dense2 = self.dense2.compute_output_shape(output_shape_dense1)
        self.output_layer.build(output_shape_dense2)
        # Call parent build method to finalize model state
        super().build(input_shape)

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

    def train_step(self, data):
        X, y = data

        # 1. Standard Forward Pass and Empirical Loss Calculation (L_k)
        with tf.GradientTape() as tape:
            y_pred = self(X, training=True)
            # Use the compiled loss (Binary Cross-Entropy)
            empirical_loss = self.compiled_loss(y, y_pred, regularization_losses=self.losses)

            # 2. Proximal Regularization Term (Omega = (mu/2) * ||w - w_t||^2)
            proximal_term = tf.constant(0.0)

            # Iterate over all trainable weights in the current model (w)
            for current_w, reference_w in zip(self.trainable_weights, self.w_t_initial):
                # Calculate ||w - w_t||^2 for each layer
                layer_prox_sum = tf.reduce_sum(tf.square(current_w - reference_w))
                proximal_term += layer_prox_sum

            proximal_loss = (self.mu / 2.0) * proximal_term

            # 3. Total Composite Loss (L_total = L_k + Omega)
            total_loss = empirical_loss + proximal_loss

        # 4. Backpropagation (Calculate gradients of L_total)
        trainable_vars = self.trainable_weights
        gradients = tape.gradient(total_loss, trainable_vars)

        # 5. Apply gradients (Optimizer update)
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))

        # --- Update explicitly defined metrics ---
        self.loss_metric.update_state(total_loss)
        self.accuracy_metric.update_state(y, y_pred)
        self.proximal_loss_tracker.update_state(proximal_loss)

        # Return a dictionary of metric results
        return {"loss": self.loss_metric.result(),
                "accuracy": self.accuracy_metric.result(),
                "proximal_loss": self.proximal_loss_tracker.result()}

    @property
    def metrics(self):
        # --- Return all explicitly defined metrics ---
        return [self.loss_metric, self.accuracy_metric, self.proximal_loss_tracker]

def create_compiled_fedprox_model(mu, global_weights_t):
    """Factory function to create and compile the FedProx MLP model."""
    model = FedProxMLP(mu, global_weights_t)
    # Call build explicitly after instantiation
    model.build((None, MLP_INPUT_DIM))
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss='binary_crossentropy' # Loss is used inside train_step
        # Metrics are handled by the custom train_step, so no need to pass here
    )
    return model
```

```python
# ============================================================
# PHASE 1 — DATA LOADING, PREPROCESSING, NON-IID PARTITIONING
# (Modified to include Volume Shift - Step 1)
# ============================================================

def load_and_preprocess_wdbc(url=DATA_URL):
    """
    Load, encode, and scale the UCI Breast Cancer WDBC dataset.
    """

    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
        ]
    ]

    data = pd.read_csv(url, header=None, names=columns)
    data = data.drop('id', axis=1)
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({'M': 1, 'B': 0})

    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)}")
    print(f"Train pool: {len(X_train_full)}, Test set: {len(X_test)}")
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


def create_non_iid_partitions(X_train, y_train, num_clients=3):
    """
    Creates Non-IID partitions with:
    1. Label Skew (via sampling strategy)
    2. Volume Shift (via VOLUME_RATIOS) - Professor's Request
    """

    df = X_train.copy()
    df['y'] = y_train.values

    benign_df = df[df['y'] == 0].reset_index(drop=True)
    malig_df = df[df['y'] == 1].reset_index(drop=True)

    # Targets: [0.20, 0.50, 0.80] malignant proportion (Label Skew)
    malignant_ratios = [0.20, 0.50, 0.80]

    # Ratios for Volume Shift (Volume Skew) - Total samples allocated [0.50, 0.30, 0.20]
    volume_ratios = VOLUME_RATIOS
    total_size = len(df)
    clients = [] # Initialized as an empty list

    print("=== NON-IID CLIENT PARTITIONS (Label Skew + Volume Shift) ===")
    for i in range(num_clients):
        # 1. Calculate N_client based on Volume Ratio (Volume Shift)
        n_client = int(total_size * volume_ratios[i])

        # 2. Calculate Label Skew within N_client (Label Skew)
        ratio = malignant_ratios[i]
        n_m = int(n_client * ratio + 0.5)
        n_b = n_client - n_m
        n_m = max(n_m, 5)
        n_b = max(n_b, 5)

        # Sample with replacement from class pools
        mal_part = malig_df.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        ben_part = benign_df.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([ben_part, mal_part], axis=0)
        client_df = client_df.sample(frac=1.0, random_state=SEED + 30*i).reset_index(drop=True)

        X_c = client_df.drop('y', axis=1)
        y_c = client_df['y']

        total_c = len(y_c)
        num_m = int(y_c.sum())

        print(f"Client {i+1}: {total_c} samples (Target: {int(total_size * volume_ratios[i])}) | Malignant ratio: {100 * num_m / total_c:.1f}%")

        clients.append((X_c, y_c))

    print("-" * 50)
    return clients
```

```python
# ============================================================
# PHASE 4 & 5 — SERVER CLASS (FedAvg/FedProx Aggregation)
# ============================================================

class FLServer:
    def __init__(self, clients, X_test, y_test, initial_global_weights):
        self.clients = clients
        self.X_test = X_test.values
        self.y_test = y_test.values
        self.history = [] # Initialized as an empty list

        # Global model is maintained as a list of weights (NumPy arrays)
        self.global_weights = [w.copy() for w in initial_global_weights]

    def sample_clients(self):
        return self.clients

    def run_training_round(self, round_num):

        selected_clients = self.sample_clients()

        all_client_updates = [] # Initialized as an empty list

        # 1. Collect Model Updates (Weights + n_k) from all clients
        for client in selected_clients:
            updated_weights, n_k = client.get_model_update_with_dp(self.global_weights)
            all_client_updates.append((updated_weights, n_k))

        # 2. Aggregation (Phase 5: Weighted FedAvg/FedProx Aggregation)
        self.global_weights = self._aggregate_mlp_fedavg(all_client_updates)

        # 3. Global Evaluation (Phase 6)

        # Temporarily create a model instance for evaluation
        eval_model = Sequential([
        Dense(16, activation='relu', input_shape=(MLP_INPUT_DIM,)),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')
        ])

        eval_model.compile(optimizer=Adam(learning_rate=0.001),loss='binary_crossentropy')

        # Trigger natural build with one forward call
        _ = eval_model(tf.zeros((1, MLP_INPUT_DIM)))
        eval_model.set_weights(self.global_weights)


        y_pred = self._get_predictions(eval_model)

        acc = accuracy_score(self.y_test, y_pred)
        sensitivity = recall_score(self.y_test, y_pred, pos_label=1)
        specificity = recall_score(self.y_test, y_pred, pos_label=0)
        f1 = f1_score(self.y_test, y_pred, pos_label=1)

        self.history.append((round_num, acc, sensitivity, specificity, f1))

        return acc, sensitivity, specificity, f1

    def _aggregate_mlp_fedavg(self, client_updates):
        """
        Weighted FedAvg/FedProx Aggregation for MLP weights (NumPy arrays).
        This correctly implements Volume Shift weighting (n_k / N).
        """

        # Calculate N: Total samples across participating clients
        total_samples = sum(n_k for w_k, n_k in client_updates)

        # Initialize zero list for accumulated weights (structure from first client)
        initial_weights = client_updates[0][0] # Access weights from the first client's update
        global_weights = [np.zeros_like(w) for w in initial_weights]

        # Aggregate the weights based on client sample size (n_k)

        for w_k, n_k in client_updates:
            # Normalization factor alpha_k (n_k / N)
            alpha_k = n_k / total_samples

            # Apply weighting and accumulate
            for i in range(len(global_weights)):
                global_weights[i] += w_k[i] * alpha_k

        return global_weights

    def _get_predictions(self, model):
        """Handles predictions for the MLP model."""
        # Use threshold 0.5 on probability output
        return (model.predict(self.X_test, verbose=0) > 0.5).astype(int).flatten()

    def run(self, n_rounds):

        print(f"\n=== STARTING PHASE B.1 & B.3 (MODEL: MLP / MU={MU}) ===")
        for r in range(1, n_rounds + 1):

            if r % 10 == 0 or r == 1 or r == n_rounds:
                print(f"\n========== ROUND {r}/{n_rounds} ==========")
                acc, sens, spec, f1 = self.run_training_round(r)
                print(f" Global metrics at round {r}:\n" \
                    f"  Accuracy = {acc:.4f}\n" \
                    f"  Sensitivity = {sens:.4f}\n" \
                    f"  Specificity = {spec:.4f}\n" \
                    f"  F1-score = {f1:.4f}")
            else:
                self.run_training_round(r)
        return self.history
```

```python
# ============================================================
# PHASE B.1 & B.3: FEDPROX IMPLEMENTATION & NON-IID VOLUME SHIFT
# Fully Patched, Cleaned, and Runnable End-to-End
# ============================================================

import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, f1_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Mean

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)

# ---------------- CONFIG ----------------
NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

MLP_INPUT_DIM = 30
MLP_EPOCHS = 5
MLP_BATCH_SIZE = 32
MLP_LEARNING_RATE = 0.001

VOLUME_RATIOS = [0.50, 0.30, 0.20]  # Unequal samples
MALIGNANT_RATIOS = [0.20, 0.50, 0.80]
N_ROUNDS = 20

MU = 0.0
USE_DP = False


# ============================================================
# FEDPROX CUSTOM MODEL
# ============================================================

class FedProxMLP(keras.Model):
    def __init__(self, mu, global_weights_t, name=None):
        super().__init__(name=name)
        self.mu = tf.constant(mu, dtype=tf.float32)
        self.anchor_weights = global_weights_t

        # Layers
        self.d1 = Dense(16, activation='relu')
        self.d2 = Dense(8, activation='relu')
        self.out = Dense(1, activation='sigmoid')

        # Metrics
        self.loss_metric = Mean(name="loss")
        self.acc_metric = keras.metrics.BinaryAccuracy(name="accuracy")
        self.prox_metric = Mean(name="proximal_loss")

    def call(self, x):
        x = self.d1(x)
        x = self.d2(x)
        return self.out(x)

    def train_step(self, data):
        X, y = data

        with tf.GradientTape() as tape:
            y_pred = self(X, training=True)
            empirical_loss = self.compiled_loss(y, y_pred)

            # Proximal penalty
            prox = tf.constant(0.0)
            for w, w_anchor in zip(self.trainable_weights, self.anchor_weights):
                prox += tf.reduce_sum(tf.square(w - w_anchor))
            prox_loss = 0.5 * self.mu * prox

            total_loss = empirical_loss + prox_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        # Metrics
        self.loss_metric.update_state(total_loss)
        self.acc_metric.update_state(y, y_pred)
        self.prox_metric.update_state(prox_loss)

        return {
            "loss": self.loss_metric.result(),
            "accuracy": self.acc_metric.result(),
            "proximal_loss": self.prox_metric.result()
        }

    @property
    def metrics(self):
        return [self.loss_metric, self.acc_metric, self.prox_metric]


def create_compiled_fedprox(mu, anchor_weights):
    model = FedProxMLP(mu, anchor_weights)
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss="binary_crossentropy"
    )
    return model


def get_initial_weights():
    """Create a temporary model to get correctly shaped weights."""
    temp = Sequential([
        Dense(16, activation='relu', input_shape=(MLP_INPUT_DIM,)),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    return temp.get_weights()


# ============================================================
# PHASE 1 — DATA
# ============================================================

def load_and_preprocess_wdbc(url=DATA_URL):
    cols = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
        ]
    ]

    df = pd.read_csv(url, header=None, names=cols).drop('id', axis=1)
    X = df.drop('diagnosis', axis=1)
    y = df['diagnosis'].map({'M': 1, 'B': 0})

    scaler = StandardScaler()
    X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total: {len(df)} | Train: {len(X_train)} | Test: {len(X_test)}")
    print("-" * 50)

    return X_train.reset_index(drop=True), y_train.reset_index(drop=True), \
           X_test.reset_index(drop=True), y_test.reset_index(drop=True)


# ============================================================
# PHASE 2 — NON-IID SPLIT
# ============================================================

def create_non_iid_partitions(X_train, y_train):
    df = X_train.copy()
    df['y'] = y_train.values

    benign = df[df['y'] == 0]
    mal = df[df['y'] == 1]

    clients = []
    total = len(df)

    print("=== NON-IID CLIENT PARTITIONS ===")

    for i in range(NUM_CLIENTS):
        volume = int(total * VOLUME_RATIOS[i])
        ratio = MALIGNANT_RATIOS[i]

        n_m = max(int(volume * ratio), 5)
        n_b = max(volume - n_m, 5)

        df_m = mal.sample(n=n_m, replace=True, random_state=SEED + i)
        df_b = benign.sample(n=n_b, replace=True, random_state=SEED + 10*i)

        cdf = pd.concat([df_m, df_b]).sample(frac=1, random_state=SEED+20*i)
        Xc = cdf.drop("y", axis=1)
        yc = cdf["y"]

        print(f"Client {i+1}: {len(yc)} samples | Malignant {100*yc.mean():.1f}%")
        clients.append((Xc, yc))

    print("-" * 50)
    return clients


# ============================================================
# PHASE 3 — CLIENT
# ============================================================

class FLClient:
    def __init__(self, cid, X, y):
        self.cid = cid
        self.X = X.values
        self.y = y.values
        self.n = len(self.X)
        print(f"[Client {cid}] Ready with {self.n} samples.")

    def train_local_model(self, global_weights):
        model = create_compiled_fedprox(MU, global_weights)

        # build before loading weights
        _ = model(self.X[:1])
        model.set_weights(global_weights)

        model.fit(self.X, self.y,
                  batch_size=MLP_BATCH_SIZE,
                  epochs=MLP_EPOCHS,
                  verbose=0)

        return model.get_weights()

    def get_update(self, global_weights):
        new = self.train_local_model(global_weights)
        return new, self.n


# ============================================================
# PHASE 4 — SERVER
# ============================================================

class FLServer:
    def __init__(self, clients, X_test, y_test, init_weights):
        self.clients = clients
        self.Xt = X_test.values
        self.yt = y_test.values
        self.history = []
        self.global_weights = init_weights

        self.eval_model = create_compiled_fedprox(0.0, init_weights)

        # build before loading weights
        _ = self.eval_model(self.Xt[:1])
        self.eval_model.set_weights(self.global_weights)

    def aggregate(self, updates):
        total = sum(n for _, n in updates)
        new_weights = [np.zeros_like(w) for w in updates[0][0]]

        for w, n in updates:
            alpha = n / total
            for i in range(len(new_weights)):
                new_weights[i] += w[i] * alpha

        return new_weights

    def run_round(self, r):
        updates = [c.get_update(self.global_weights) for c in self.clients]

        self.global_weights = self.aggregate(updates)

        self.eval_model.set_weights(self.global_weights)
        preds = (self.eval_model.predict(self.Xt, verbose=0) > 0.5).astype(int).flatten()

        acc = accuracy_score(self.yt, preds)
        sens = recall_score(self.yt, preds, pos_label=1)
        spec = recall_score(self.yt, preds, pos_label=0)
        f1 = f1_score(self.yt, preds)

        self.history.append((r, acc, sens, spec, f1))
        return acc, sens, spec, f1

    def run(self, rounds):
        print(f"\n=== TRAINING START (MU={MU}) ===")
        for r in range(1, rounds+1):
            if r == 1 or r % 5 == 0 or r == rounds:
                print(f"\n--- Round {r}/{rounds} ---")
            acc, sens, spec, f1 = self.run_round(r)
            if r == 1 or r % 5 == 0 or r == rounds:
                print(f"ACC={acc:.4f}  SEN={sens:.4f}  SPEC={spec:.4f}  F1={f1:.4f}")
        return self.history


# ============================================================
# EXECUTION
# ============================================================

X_train, y_train, X_test, y_test = load_and_preprocess_wdbc()
client_parts = create_non_iid_partitions(X_train, y_train)

initial_global_weights = get_initial_weights()

# FedAvg
MU = 0.0
clients_fedavg = [FLClient(i+1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]
server_fedavg = FLServer(clients_fedavg, X_test, y_test, initial_global_weights)
history_fedavg = server_fedavg.run(N_ROUNDS)

# FedProx
MU = 0.01
clients_fedprox = [FLClient(i+1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]
server_fedprox = FLServer(clients_fedprox, X_test, y_test, initial_global_weights)
history_fedprox = server_fedprox.run(N_ROUNDS)

print("\n=== FINAL RESULTS ===")
print(f"FedAvg Final: {history_fedavg[-1]}")
print(f"FedProx Final: {history_fedprox[-1]}")
```

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, f1_score
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras.metrics import Mean

# Reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------- CONFIGURATION ----------------
NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

# Model Parameters
MLP_INPUT_DIM = 30
MLP_EPOCHS = 10
MLP_BATCH_SIZE = 32
MLP_LEARNING_RATE = 0.001

# Non-IID Settings
VOLUME_RATIOS = [0.50, 0.30, 0.20]
N_ROUNDS = 20

FEATURE_SHIFT_CLIENT_ID = 3
FEATURE_SHIFT_FACTOR = 5.0

MU = 0.0
USE_DP = False


# ============================================================
# BASELINE MLP ARCHITECTURE (used for eval + init weights)
# ============================================================
def build_model(input_dim):
    model = Sequential([
        Input((input_dim,)),
        Dense(16, activation="relu"),
        Dense(8, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    # trigger natural build
    _ = model(tf.zeros((1, input_dim)))
    return model


# ============================================================
# FEDPROX MLP MODEL
# ============================================================
class FedProxMLP(keras.Model):
    def __init__(self, mu, global_weights_t, name=None):
        super().__init__(name=name)
        self.mu = tf.constant(mu, dtype=tf.float32)
        self.w_t_initial = global_weights_t  # list of numpy arrays (anchor weights)

        # define layers (must match build_model/get_initial_weights)
        self.dense1 = Dense(16, activation="relu")
        self.dense2 = Dense(8, activation="relu")
        self.output_layer = Dense(1, activation="sigmoid")

        # metrics
        self.loss_metric = Mean(name="loss")
        self.accuracy_metric = keras.metrics.BinaryAccuracy(name="accuracy")
        self.proximal_loss_tracker = Mean(name="proximal_loss")

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

    def train_step(self, data):
        X, y = data

        with tf.GradientTape() as tape:
            y_pred = self(X, training=True)
            empirical_loss = self.compiled_loss(y, y_pred, regularization_losses=self.losses)

            proximal_term = tf.constant(0.0, dtype=self.mu.dtype)
            for current_w, reference_w in zip(self.trainable_weights, self.w_t_initial):
                ref_t = tf.convert_to_tensor(reference_w, dtype=current_w.dtype)
                proximal_term += tf.reduce_sum(tf.square(current_w - ref_t))

            proximal_loss = (self.mu / 2.0) * proximal_term
            total_loss = empirical_loss + proximal_loss

        gradients = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_weights))

        # update metrics
        self.loss_metric.update_state(total_loss)
        self.accuracy_metric.update_state(y, y_pred)
        self.proximal_loss_tracker.update_state(proximal_loss)

        return {
            "loss": self.loss_metric.result(),
            "accuracy": self.accuracy_metric.result(),
            "proximal_loss": self.proximal_loss_tracker.result()
        }

    @property
    def metrics(self):
        return [self.loss_metric, self.accuracy_metric, self.proximal_loss_tracker]


def create_compiled_fedprox_model(mu, global_weights_t):
    """
    For mu > 0: FedProx custom model using global_weights_t as anchor.
    """
    model = FedProxMLP(mu, global_weights_t)
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss="binary_crossentropy"
    )
    _ = model(tf.zeros((1, MLP_INPUT_DIM)))  # trigger build
    model.set_weights(global_weights_t)
    return model


# ============================================================
# INITIAL WEIGHTS: build temp identical MLP and return its weights
# ============================================================
def get_initial_weights(input_dim):
    """
    Create a temporary model with identical architecture to obtain correctly-shaped init weights.
    """
    temp_model = build_model(input_dim)
    return temp_model.get_weights()


# ============================================================
# DATA LOADING, GLOBAL SCALING, and bookkeeping for feature shift
# ============================================================
def load_and_preprocess_wdbc(url=DATA_URL):
    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ["mean", "se", "worst"]
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry',
            'fractal_dimension'
        ]
    ]
    data = pd.read_csv(url, header=None, names=columns)
    data = data.drop('id', axis=1)
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({"M": 1, "B": 0})

    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    # Save feature index for 'radius_mean' (used for targeted shift)
    global FEATURE_TO_SHIFT_INDEX
    FEATURE_TO_SHIFT_INDEX = X.columns.get_loc('radius_mean')

    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)} | Train pool: {len(X_train_full)} | Test set: {len(X_test)}")
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


# ============================================================
# FEATURE SHIFT: apply to chosen client only
# ============================================================
def apply_feature_shift(X_client, client_id):
    """
    Applies shift only to column 'radius_mean' (preserves other columns).
    X_client is standardized data (mean=0, std=1)
    """
    if client_id == FEATURE_SHIFT_CLIENT_ID:
        X_client = X_client.copy()
        X_client.iloc[:, FEATURE_TO_SHIFT_INDEX] = (
            X_client.iloc[:, FEATURE_TO_SHIFT_INDEX] + FEATURE_SHIFT_FACTOR
        )
        print(f"  Client {client_id} radius_mean shifted by +{FEATURE_SHIFT_FACTOR} std dev.")
    return X_client


# ============================================================
# NON-IID PARTITIONING (Label skew + Volume shift + Feature shift)
# ============================================================
def create_non_iid_partitions(X_train, y_train, num_clients):
    df = X_train.copy()
    df["y"] = y_train

    benign = df[df["y"] == 0].reset_index(drop=True)
    malignant = df[df["y"] == 1].reset_index(drop=True)

    malignant_ratios = [0.20, 0.50, 0.80]
    volume_ratios = VOLUME_RATIOS
    total = len(df)

    clients = []

    print("\n=== NON-IID PARTITIONS ===")
    for i in range(num_clients):
        cid = i + 1
        size = int(volume_ratios[i] * total)

        ratio = malignant_ratios[i]
        n_m = max(5, int(size * ratio))
        n_b = max(5, size - n_m)

        part_m = malignant.sample(n_m, replace=True, random_state=SEED + 10*i)
        part_b = benign.sample(n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([part_b, part_m]).sample(frac=1, random_state=SEED + 30*i)

        X_c = client_df.drop("y", axis=1)
        y_c = client_df["y"]

        X_c = apply_feature_shift(X_c, cid)

        print(f"Client {cid}: {len(X_c)} samples | Malignant={y_c.mean()*100:.1f}%")

        clients.append((X_c, y_c))

    print("--------------------------------------")
    return clients


# ============================================================
# CLIENT CLASS (FedAvg + FedProx)
# ============================================================
class FLClient:
    def __init__(self, cid, X, y):
        self.cid = cid
        self.X = X.values.astype(np.float32)
        self.y = y.values.astype(np.float32)
        self.input_dim = X.shape[1]
        self.n = len(self.X)
        print(f"[Client {cid}] Ready ({self.n} samples)")

    def local_train(self, global_weights, epochs=1, batch_size=32, mu=0.0):
        if mu == 0.0:
            # FedAvg: plain MLP
            model = build_model(self.input_dim)
            model.set_weights(global_weights)
        else:
            # FedProx: custom model with proximal term
            model = create_compiled_fedprox_model(mu, global_weights)

        model.fit(self.X, self.y, epochs=epochs, batch_size=batch_size, verbose=0)
        return model

    def get_model_update_with_dp(self, global_weights_t, mu):
        model = self.local_train(
            global_weights_t,
            epochs=MLP_EPOCHS,
            batch_size=MLP_BATCH_SIZE,
            mu=mu
        )
        updated_weights = model.get_weights()

        if USE_DP:
            # DP placeholder (not implemented)
            pass

        return updated_weights, self.n


# ============================================================
# SERVER CLASS
# ============================================================
class FLServer:
    def __init__(self, clients, X_test, y_test, initial_weights):
        self.clients = clients
        self.X_test = X_test.values.astype(np.float32)
        self.y_test = y_test.values.astype(np.float32)
        self.global_weights = [w.copy() for w in initial_weights]

        self.input_dim = self.X_test.shape[1]
        self.eval_model = build_model(self.input_dim)
        self.eval_model.set_weights(self.global_weights)

    def aggregate(self, client_results):
        total = sum(sz for _, sz in client_results)
        new_weights = []

        for layer_idx in range(len(self.global_weights)):
            layer_sum = np.zeros_like(self.global_weights[layer_idx])
            for weights, size in client_results:
                layer_sum += (size / total) * weights[layer_idx]
            new_weights.append(layer_sum)
        return new_weights

    def evaluate(self):
        self.eval_model.set_weights(self.global_weights)
        loss, acc = self.eval_model.evaluate(self.X_test, self.y_test, verbose=0)
        return acc

    def run(self, rounds, mu=0.0):
        history = []

        print(f"\n===== TRAINING (MU={mu}) =====")
        for r in range(1, rounds + 1):
            print(f"--- Round {r}/{rounds} ---")
            results = []

            for client in self.clients:
                w, sz = client.get_model_update_with_dp(self.global_weights, mu)
                results.append((w, sz))

            self.global_weights = self.aggregate(results)
            test_acc = self.evaluate()
            print(f"Global Test Accuracy: {test_acc:.4f}")

            history.append(test_acc)

        return history


# ============================================================
# RUN EXPERIMENT
# ============================================================
if __name__ == "__main__":
    # --- Data Preparation ---
    X_train_full, y_train_full, X_test, y_test = load_and_preprocess_wdbc()
    client_parts = create_non_iid_partitions(X_train_full, y_train_full, NUM_CLIENTS)

    # --- 1. Generate Initial Global Weights Structure (Reusable) ---
    initial_global_weights_zero = get_initial_weights(X_train_full.shape[1])

    # --- 2. Experiment 1: FedAvg Baseline (MU = 0.0) ---
    MU_FEDAVG = 0.0
    clients_fedavg = [FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]
    server_fedavg = FLServer(clients_fedavg, X_test, y_test, initial_global_weights_zero)
    history_fedavg = server_fedavg.run(N_ROUNDS, mu=MU_FEDAVG)

    # --- 3. Experiment 2: FedProx (MU = 0.1) ---
    MU_FEDPROX = 0.1
    clients_fedprox = [FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]
    server_fedprox = FLServer(clients_fedprox, X_test, y_test, initial_global_weights_zero)
    history_fedprox = server_fedprox.run(N_ROUNDS, mu=MU_FEDPROX)

    # --- 4. Final Comparison ---
    print("\n=== FINAL RESULTS ===")
    print("FedAvg Final Accuracy:", history_fedavg[-1])
    print("FedProx Final Accuracy:", history_fedprox[-1])
```

```python
# ============================================================
# PHASE B.1 & B.3: FEDPROX IMPLEMENTATION + FEATURE SHIFT
# ============================================================

import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, f1_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Mean

# ----------------- CONFIG & SEED -----------------
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

MLP_INPUT_DIM = 30
MLP_EPOCHS = 10
MLP_BATCH_SIZE = 32
MLP_LEARNING_RATE = 0.001
N_ROUNDS = 20   # you can set 50 later if you want

# Volume + label skew
VOLUME_RATIOS = [0.50, 0.30, 0.20]
MALIGNANT_RATIOS = [0.20, 0.50, 0.80]

# Feature shift on radius_mean for client 3
FEATURE_SHIFT_CLIENT_ID = 3
FEATURE_SHIFT_FACTOR = 5.0   # +5 std dev shift

USE_DP = False   # placeholder for DP noise later

# Will be set when we load data
RADIUS_MEAN_IDX = None


# ============================================================
# FEDPROX MLP MODEL
# ============================================================

class FedProxMLP(keras.Model):
    """
    Keras model with FedProx objective:
        L_k(w) + (mu/2) * ||w - w_t||^2
    """
    def __init__(self, mu, global_weights_t, name=None):
        super().__init__(name=name)
        self.mu = tf.constant(mu, dtype=tf.float32)

        # Anchor weights w_t (as numpy arrays)
        self.w_t_anchor = None
        if global_weights_t is not None:
            self.w_t_anchor = [np.array(w) for w in global_weights_t]

        # Architecture
        self.dense1 = Dense(16, activation="relu")
        self.dense2 = Dense(8, activation="relu")
        self.output_layer = Dense(1, activation="sigmoid")

        # Metrics
        self.loss_metric = Mean(name="loss")
        self.accuracy_metric = keras.metrics.BinaryAccuracy(name="accuracy")
        self.proximal_loss_tracker = Mean(name="proximal_loss")

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

    def train_step(self, data):
        X, y = data
        with tf.GradientTape() as tape:
            y_pred = self(X, training=True)
            # cross-entropy
            empirical_loss = self.compiled_loss(y, y_pred)

            # FedProx term
            proximal_loss = tf.constant(0.0, dtype=tf.float32)
            if self.w_t_anchor is not None and float(self.mu.numpy()) > 0.0:
                prox_term = tf.constant(0.0, dtype=tf.float32)
                for w, w_t in zip(self.trainable_weights, self.w_t_anchor):
                    prox_term += tf.reduce_sum(
                        tf.square(w - tf.convert_to_tensor(w_t, dtype=w.dtype))
                    )
                proximal_loss = (self.mu / 2.0) * prox_term

            total_loss = empirical_loss + proximal_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        # Update metrics
        self.loss_metric.update_state(total_loss)
        self.accuracy_metric.update_state(y, y_pred)
        self.proximal_loss_tracker.update_state(proximal_loss)

        return {
            "loss": self.loss_metric.result(),
            "accuracy": self.accuracy_metric.result(),
            "proximal_loss": self.proximal_loss_tracker.result()
        }

    @property
    def metrics(self):
        return [self.loss_metric, self.accuracy_metric, self.proximal_loss_tracker]


def create_fedprox_model(mu, global_weights_t):
    """
    Create a FedProxMLP and (optionally) load global weights.
    Used for BOTH FedAvg (mu=0) and FedProx (mu>0).
    """
    model = FedProxMLP(mu, global_weights_t)
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss="binary_crossentropy"
    )
    # build variables
    _ = model(tf.zeros((1, MLP_INPUT_DIM)))
    if global_weights_t is not None:
        model.set_weights(global_weights_t)
    return model


def get_initial_weights():
    """
    Create a FedProxMLP once (mu=0, no anchor) and return its random initial weights.
    This defines the weight structure for the entire experiment.
    """
    temp_model = create_fedprox_model(mu=0.0, global_weights_t=None)
    return temp_model.get_weights()


# ============================================================
# DATA LOADING + CENTRAL SPLIT
# ============================================================

def load_and_preprocess_wdbc():
    global RADIUS_MEAN_IDX

    feature_cols = [
        f"{feat}_{stat}"
        for stat in ["mean", "se", "worst"]
        for feat in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry',
            'fractal_dimension'
        ]
    ]
    cols = ["id", "diagnosis"] + feature_cols

    data = pd.read_csv(DATA_URL, header=None, names=cols)
    X = data[feature_cols]
    y = data["diagnosis"].map({"M": 1, "B": 0})

    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_cols)

    # index for radius_mean (used for targeted feature shift)
    RADIUS_MEAN_IDX = X_scaled.columns.get_loc("radius_mean")

    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled,
        y,
        test_size=TEST_SPLIT_RATIO,
        stratify=y,
        random_state=SEED
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)} | Train={len(X_train)}, Test={len(X_test)}")
    print("-" * 50)

    return (
        X_train.reset_index(drop=True),
        y_train.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


# ============================================================
# NON-IID PARTITIONING (VOLUME SKEW + LABEL SKEW + FEATURE SHIFT)
# ============================================================

def create_non_iid_partitions(X_train, y_train):
    df = X_train.copy()
    df["y"] = y_train.values

    benign = df[df["y"] == 0].reset_index(drop=True)
    malignant = df[df["y"] == 1].reset_index(drop=True)

    clients = []
    total = len(df)

    print("\n=== NON-IID CLIENT PARTITIONS (Volume + Label Skew + Feature Shift) ===")
    for i in range(NUM_CLIENTS):
        cid = i + 1
        size = int(VOLUME_RATIOS[i] * total)

        ratio = MALIGNANT_RATIOS[i]
        n_m = max(5, int(size * ratio))
        n_b = max(5, size - n_m)

        part_m = malignant.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        part_b = benign.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([part_b, part_m]).sample(
            frac=1.0, random_state=SEED + 30*i
        )

        X_c = client_df.drop("y", axis=1)
        y_c = client_df["y"]

        # Apply feature shift only to chosen client
        if cid == FEATURE_SHIFT_CLIENT_ID:
            X_c = X_c.copy()
            X_c.iloc[:, RADIUS_MEAN_IDX] = X_c.iloc[:, RADIUS_MEAN_IDX] + FEATURE_SHIFT_FACTOR
            print(f"  >> Client {cid}: radius_mean shifted by +{FEATURE_SHIFT_FACTOR} std")

        print(f"Client {cid}: {len(X_c)} samples | Malignant={100*y_c.mean():.1f}%")
        clients.append((X_c.reset_index(drop=True), y_c.reset_index(drop=True)))

    print("-" * 50)
    return clients


# ============================================================
# CLIENT
# ============================================================

class FLClient:
    def __init__(self, cid, X_local, y_local):
        self.cid = cid
        self.X = X_local.values.astype(np.float32)
        self.y = y_local.values.astype(np.float32)
        self.n_samples = len(self.X)
        print(f"[Client {cid}] initialized with {self.n_samples} samples.")

    def get_model_update(self, global_weights, mu):
        """
        Train local FedProx (or FedAvg if mu=0) model and return updated weights + sample count.
        """
        local_model = create_fedprox_model(mu=mu, global_weights_t=global_weights)
        local_model.fit(
            self.X,
            self.y,
            epochs=MLP_EPOCHS,
            batch_size=MLP_BATCH_SIZE,
            verbose=0
        )
        updated_weights = local_model.get_weights()

        if USE_DP:
            # TODO: add DP noise here if needed
            pass

        return updated_weights, self.n_samples


# ============================================================
# SERVER
# ============================================================

class FLServer:
    def __init__(self, clients, X_test, y_test, initial_global_weights):
        self.clients = clients
        self.X_test = X_test.values.astype(np.float32)
        self.y_test = y_test.values.astype(np.float32)
        self.global_weights = [w.copy() for w in initial_global_weights]

        # Evaluation model (no proximal term used for eval)
        self.eval_model = FedProxMLP(mu=0.0, global_weights_t=None)
        _ = self.eval_model(tf.zeros((1, MLP_INPUT_DIM)))  # build weights

    def aggregate(self, client_updates):
        total = sum(n for _, n in client_updates)
        new_weights = [np.zeros_like(w) for w in self.global_weights]
        for w, n in client_updates:
            alpha = n / total
            for i in range(len(new_weights)):
                new_weights[i] += alpha * w[i]
        return new_weights

    def evaluate(self, round_num):
        self.eval_model.set_weights(self.global_weights)
        probs = self.eval_model(self.X_test, training=False).numpy().reshape(-1)
        preds = (probs > 0.5).astype(int)

        acc = accuracy_score(self.y_test, preds)
        sens = recall_score(self.y_test, preds, pos_label=1)
        spec = recall_score(self.y_test, preds, pos_label=0)
        f1 = f1_score(self.y_test, preds)

        print(f"Round {round_num}: "
              f"Acc={acc:.4f} | Sens={sens:.4f} | Spec={spec:.4f} | F1={f1:.4f}")
        return acc, sens, spec, f1

    def run(self, n_rounds, mu):
        history = []
        print(f"\n=== START TRAINING (MU={mu}) ===")
        for r in range(1, n_rounds + 1):
            # Collect client updates
            updates = []
            for c in self.clients:
                w, n = c.get_model_update(self.global_weights, mu)
                updates.append((w, n))

            # FedAvg aggregation (weights by n_k)
            self.global_weights = self.aggregate(updates)

            # Evaluate on central test set
            metrics = self.evaluate(r)
            history.append((r,) + metrics)

        return history


# ============================================================
# MAIN EXPERIMENT
# ============================================================

def main():
    # 1) Load and split data
    X_train, y_train, X_test, y_test = load_and_preprocess_wdbc()

    # 2) Create Non-IID client partitions (volume + label skew + feature shift)
    client_partitions = create_non_iid_partitions(X_train, y_train)

    # 3) Initialize random global weights once (reused for both experiments)
    initial_global_weights = get_initial_weights()

    # ---------------- FedAvg baseline (mu = 0.0) ----------------
    mu_fedavg = 0.0
    print("\n######## FEDAVG (MU=0.0) ########")
    clients_fedavg = [
        FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_partitions)
    ]
    server_fedavg = FLServer(clients_fedavg, X_test, y_test, initial_global_weights)
    history_fedavg = server_fedavg.run(N_ROUNDS, mu=mu_fedavg)
    print("\n=== FEDAVG COMPLETE (MU=0.0) ===")
    print("Final round metrics:", history_fedavg[-1])

    # ---------------- FedProx (mu > 0) ----------------
    mu_fedprox = 0.01
    print("\n######## FEDPROX (MU=0.01) ########")
    clients_fedprox = [
        FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_partitions)
    ]
    server_fedprox = FLServer(clients_fedprox, X_test, y_test, initial_global_weights)
    history_fedprox = server_fedprox.run(N_ROUNDS, mu=mu_fedprox)
    print("\n=== FEDPROX COMPLETE (MU=0.01) ===")
    print("Final round metrics:", history_fedprox[-1])


# In Colab, this will run when you execute the cell
if __name__ == "__main__":
    main()
```

```python
# ================================================================
#  FINAL CLEAN + PATCHED FEDERATED LEARNING SCRIPT
#  - Non-IID partitions via Dirichlet
#  - FedAvg + FedProx
#  - Matching MLP architectures (NO WEIGHT MISMATCH ISSUES)
# ================================================================

import numpy as np
import random

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, f1_score

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

# ----------------- SEEDING -----------------
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ================================================================
# 1) LOAD + PREPROCESS BREAST CANCER DATA
# ================================================================
data = load_breast_cancer()
X = data.data
y = data.target

# Standard scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Central train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=SEED, stratify=y
)

print("=== DATA LOADED ===")
print(f"Train={len(X_train)}, Test={len(X_test)}\n")

# ================================================================
# 2) NON-IID PARTITIONING (Dirichlet + optional feature shift)
# ================================================================
def create_non_iid_partitions(X, y, n_clients=3, alpha=0.3):
    """
    Dirichlet label skew across clients.
    NOTE: Small alpha (0.3) can create very imbalanced clients (that is expected).
    """
    np.random.seed(SEED)
    y = np.array(y)

    # indices per class
    idx_by_class = {c: np.where(y == c)[0] for c in np.unique(y)}
    client_indices = [[] for _ in range(n_clients)]

    for c, idxs in idx_by_class.items():
        np.random.shuffle(idxs)
        proportions = np.random.dirichlet([alpha] * n_clients)
        split = (proportions * len(idxs)).astype(int)

        # Fix rounding so total matches
        while split.sum() < len(idxs):
            split[np.argmax(proportions)] += 1

        parts = np.split(idxs, np.cumsum(split)[:-1])
        for cid, part in enumerate(parts):
            client_indices[cid].extend(part)

    parts = []
    print("=== NON-IID PARTITIONS ===")
    for cid, idxs in enumerate(client_indices):
        idxs = np.array(idxs, dtype=int)
        Xc = X[idxs]
        yc = y[idxs]

        mal_rate = yc.mean() * 100
        print(f"Client {cid+1}: {len(idxs)} samples | Malignant={mal_rate:.1f}%")

        # Inject distribution shift ONLY to client 3 (cid == 2)
        if cid == 2 and len(Xc) > 0:
            print("  >> Client 3: Applied feature shift (+5.0 std)")
            Xc = Xc + 5.0 * Xc.std(axis=0)

        parts.append((Xc, yc))
    print("--------------------------------------")
    return parts

client_parts = create_non_iid_partitions(X_train, y_train, n_clients=3, alpha=0.3)

# ================================================================
# 3) BUILD MLP (shared architecture for all clients)
# ================================================================
def build_model(input_dim):
    model = Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(16, activation="relu"),
        layers.Dense(8, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

# ================================================================
# 4) INITIAL GLOBAL WEIGHTS (must match EXACT architecture)
# ================================================================
def get_initial_weights(input_dim):
    print(">> Building temp MLP for initial weights...")
    temp = Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(16, activation="relu"),
        layers.Dense(8, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    temp.build((None, input_dim))
    w = temp.get_weights()
    print(f">> Initial weights created: {len(w)} tensors\n")
    return w

initial_global_weights = get_initial_weights(X_train.shape[1])

# ================================================================
# 5) CLIENT CLASS (FedAvg + optional FedProx)
# ================================================================
class FLClient:
    def __init__(self, cid, X, y):
        self.cid = cid
        self.X = X
        self.y = y
        self.input_dim = X.shape[1]
        print(f"[Client {cid}] Ready ({len(X)} samples)")

    def local_train(self, global_weights, epochs=1, batch_size=32, mu=0.0):
        model = build_model(self.input_dim)
        model.set_weights(global_weights)

        # FedProx: modify loss if mu > 0
        if mu > 0.0:
            global_vars = [tf.convert_to_tensor(w, dtype=tf.float32) for w in global_weights]

            def fedprox_loss(global_vars):
                def loss_fn(y_true, y_pred):
                    base = keras.losses.binary_crossentropy(y_true, y_pred)
                    prox = 0.0
                    for w, gw in zip(model.trainable_weights, global_vars):
                        prox += tf.nn.l2_loss(w - gw)
                    return base + mu * prox
                return loss_fn

            model.compile(
                optimizer="adam",
                loss=fedprox_loss(global_vars),
                metrics=["accuracy"]
            )

        model.fit(
            self.X,
            self.y,
            batch_size=batch_size,
            epochs=epochs,
            verbose=0
        )
        return model.get_weights(), len(self.X)

# ================================================================
# 6) SERVER CLASS
# ================================================================
class FLServer:
    def __init__(self, clients, X_test, y_test, initial_weights):
        self.clients = clients
        self.X_test = X_test
        self.y_test = y_test

        # deep copy of global weights
        self.global_weights = [w.copy() for w in initial_weights]
        self.input_dim = X_test.shape[1]

    def aggregate(self, client_results):
        total = sum(sz for _, sz in client_results)
        new_weights = []

        for layer_idx in range(len(self.global_weights)):
            layer_sum = np.zeros_like(self.global_weights[layer_idx])
            for weights, size in client_results:
                layer_sum += (size / total) * weights[layer_idx]
            new_weights.append(layer_sum)
        return new_weights

    def evaluate(self):
        model = build_model(self.input_dim)
        model.set_weights(self.global_weights)
        y_prob = model.predict(self.X_test, verbose=0).ravel()
        y_pred = (y_prob > 0.5).astype(int)

        acc = accuracy_score(self.y_test, y_pred)
        sens = recall_score(self.y_test, y_pred, pos_label=1)
        spec = recall_score(self.y_test, y_pred, pos_label=0)
        f1 = f1_score(self.y_test, y_pred)

        print(f"Global Test -> Acc={acc:.4f} | Sens={sens:.4f} | Spec={spec:.4f} | F1={f1:.4f}")
        return acc

    def run(self, rounds=5, mu=0.0):
        history = []
        print(f"\n=== TRAINING START (mu={mu}) ===")
        for r in range(1, rounds + 1):
            print(f"\n--- ROUND {r} ---")
            results = []
            for client in self.clients:
                w, sz = client.local_train(self.global_weights, mu=mu)
                results.append((w, sz))

            self.global_weights = self.aggregate(results)
            test_acc = self.evaluate()
            history.append(test_acc)
        return history

# ================================================================
# 7) RUN FEDAVG
# ================================================================
print("=== RUNNING FEDAVG (mu = 0.0) ===")
clients_fedavg = [FLClient(i+1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]

server_fedavg = FLServer(clients_fedavg, X_test, y_test, initial_global_weights)
hist_fedavg = server_fedavg.run(rounds=5, mu=0.0)

# ================================================================
# 8) RUN FEDPROX
# ================================================================
print("\n=== RUNNING FEDPROX (mu = 0.1) ===")
MU = 0.1
clients_fedprox = [FLClient(i+1, Xc, yc) for i, (Xc, yc) in enumerate(client_parts)]

server_fedprox = FLServer(clients_fedprox, X_test, y_test, initial_global_weights)
hist_fedprox = server_fedprox.run(rounds=5, mu=MU)

print("\n=== DONE ===")
print("FedAvg history:", hist_fedavg)
print("FedProx history:", hist_fedprox)
```

```python
# ============================================================
# PHASE B.4: IMPLEMENTING FEATURE DISTRIBUTION SHIFT
# - Introduces artificial domain shift to test FedProx robustness.
# - Fixed: initial weights, model build before set_weights,
#   completed list/variable initializations and aggregation.
# ============================================================

import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, f1_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Mean

# For reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------- CONFIGURATION ----------------
NUM_CLIENTS = 3
TEST_SPLIT_RATIO = 0.2
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

# Experiment Parameters
MLP_INPUT_DIM = 30
MLP_EPOCHS = 10
MLP_BATCH_SIZE = 32
MLP_LEARNING_RATE = 0.001

# Non-IID Volume Shift Parameters
VOLUME_RATIOS = [0.50, 0.30, 0.20]  # C1=50%, C2=30%, C3=20%
N_ROUNDS = 20  # rounds for testing
FEATURE_SHIFT_CLIENT_ID = 3  # Client index (1-based) to shift
FEATURE_SHIFT_FACTOR = 5.0  # shift magnitude (std devs)

# Global MU placeholder
MU = 0.0
USE_DP = False  # placeholder; not implemented here


# ============================================================
# UTILITY: FedProx MLP (subclassed model with proximal loss)
# ============================================================
class FedProxMLP(keras.Model):
    def __init__(self, mu, global_weights_t, name=None):
        super().__init__(name=name)
        self.mu = tf.constant(mu, dtype=tf.float32)
        self.w_t_initial = global_weights_t

        # define layers (must match get_initial_weights)
        self.dense1 = Dense(16, activation="relu")
        self.dense2 = Dense(8, activation="relu")
        self.output_layer = Dense(1, activation="sigmoid")

        # metrics
        self.loss_metric = Mean(name="loss")
        self.accuracy_metric = keras.metrics.BinaryAccuracy(name="accuracy")
        self.proximal_loss_tracker = Mean(name="proximal_loss")

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

    def train_step(self, data):
        X, y = data

        with tf.GradientTape() as tape:
            y_pred = self(X, training=True)
            empirical_loss = self.compiled_loss(y, y_pred, regularization_losses=self.losses)

            proximal_term = tf.constant(0.0, dtype=self.mu.dtype)
            # convert anchor weights to tensors with matching dtype where needed
            for current_w, reference_w in zip(self.trainable_weights, self.w_t_initial):
                ref_t = tf.convert_to_tensor(reference_w, dtype=current_w.dtype)
                proximal_term += tf.reduce_sum(tf.square(current_w - ref_t))

            proximal_loss = (self.mu / 2.0) * proximal_term
            total_loss = empirical_loss + proximal_loss

        gradients = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_weights))

        # update metrics
        self.loss_metric.update_state(total_loss)
        self.accuracy_metric.update_state(y, y_pred)
        self.proximal_loss_tracker.update_state(proximal_loss)

        return {
            "loss": self.loss_metric.result(),
            "accuracy": self.accuracy_metric.result(),
            "proximal_loss": self.proximal_loss_tracker.result()
        }

    @property
    def metrics(self):
        return [self.loss_metric, self.accuracy_metric, self.proximal_loss_tracker]


def create_compiled_fedprox_model(mu, global_weights_t):
    model = FedProxMLP(mu, global_weights_t)
    model.compile(
        optimizer=Adam(learning_rate=MLP_LEARNING_RATE),
        loss="binary_crossentropy"
    )
    return model


# ============================================================
# INITIAL WEIGHTS: build temp identical MLP and return its weights
# (Option A: use same architecture initialization)
# ============================================================
def get_initial_weights(input_dim):
    """
    Create a temporary model with identical architecture to obtain correctly-shaped init weights.
    """
    temp_model = Sequential([
        Input(shape=(input_dim,)),
        Dense(16, activation="relu"),
        Dense(8, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    # Force building weights by calling build or a dummy pass
    temp_model.build((None, input_dim))
    # Return numpy arrays of weights
    return temp_model.get_weights()


# ============================================================
# DATA LOADING, GLOBAL SCALING, and bookkeeping for feature shift
# ============================================================
def load_and_preprocess_wdbc(url=DATA_URL):
    columns = ['id', 'diagnosis'] + [
        f'{feature}_{stat}'
        for stat in ['mean', 'se', 'worst']
        for feature in [
            'radius', 'texture', 'perimeter', 'area', 'smoothness',
            'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
        ]
    ]
    data = pd.read_csv(url, header=None, names=columns)
    data = data.drop('id', axis=1)
    X = data.drop('diagnosis', axis=1)
    y = data['diagnosis'].map({'M': 1, 'B': 0})

    # Fit scaler globally (important for consistent feature shift)
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    # Save feature index for 'radius_mean' (used for targeted shift)
    global FEATURE_TO_SHIFT_INDEX, GLOBAL_MEAN, GLOBAL_STD
    FEATURE_TO_SHIFT_INDEX = X.columns.get_loc('radius_mean')
    GLOBAL_MEAN = scaler.mean_
    GLOBAL_STD = scaler.scale_

    X_train_full, X_test, y_train_full, y_test = train_test_split(
        X_scaled, y,
        test_size=TEST_SPLIT_RATIO,
        random_state=SEED,
        stratify=y
    )

    print("=== PHASE 1: CENTRAL DATA SPLIT ===")
    print(f"Total samples: {len(data)} | Train pool: {len(X_train_full)} | Test set: {len(X_test)}")
    print("-" * 50)

    return (
        X_train_full.reset_index(drop=True),
        y_train_full.reset_index(drop=True),
        X_test.reset_index(drop=True),
        y_test.reset_index(drop=True)
    )


# ============================================================
# FEATURE SHIFT: apply to chosen client only
# ============================================================
def apply_feature_shift(X_client, client_id):
    """
    Applies shift only to column 'radius_mean' (preserves other columns).
    X_client is standardized data (mean=0, std=1)
    """
    if client_id == FEATURE_SHIFT_CLIENT_ID:
        # add FEATURE_SHIFT_FACTOR standard deviations to radius_mean column only
        X_client = X_client.copy()
        X_client.iloc[:, FEATURE_TO_SHIFT_INDEX] = X_client.iloc[:, FEATURE_TO_SHIFT_INDEX] + FEATURE_SHIFT_FACTOR
        print(f"  Client {client_id} radius_mean shifted by +{FEATURE_SHIFT_FACTOR} std dev.")
    return X_client


# ============================================================
# NON-IID PARTITIONING (Label skew + Volume shift + Feature shift)
# ============================================================
def create_non_iid_partitions(X_train, y_train, num_clients=3):
    df = X_train.copy()
    df['y'] = y_train.values

    benign_df = df[df['y'] == 0].reset_index(drop=True)
    malig_df = df[df['y'] == 1].reset_index(drop=True)

    malignant_ratios = [0.20, 0.50, 0.80]
    volume_ratios = VOLUME_RATIOS
    total_size = len(df)

    clients = []  # <-- fixed

    print("=== NON-IID CLIENT PARTITIONS (Label + Volume + FEATURE Shift) ===")
    for i in range(num_clients):
        client_id = i + 1
        n_client = int(total_size * volume_ratios[i])

        ratio = malignant_ratios[i]
        n_m = int(n_client * ratio + 0.5)
        n_b = n_client - n_m
        n_m = max(n_m, 5)
        n_b = max(n_b, 5)

        mal_part = malig_df.sample(n=n_m, replace=True, random_state=SEED + 10*i)
        ben_part = benign_df.sample(n=n_b, replace=True, random_state=SEED + 20*i)

        client_df = pd.concat([ben_part, mal_part], axis=0)
        client_df = client_df.sample(frac=1.0, random_state=SEED + 30*i).reset_index(drop=True)

        X_c = client_df.drop('y', axis=1)
        y_c = client_df['y']

        # apply feature shift (only modifies radius_mean column)
        X_c = apply_feature_shift(X_c, client_id)

        total_c = len(y_c)
        num_m = int(y_c.sum())

        print(f"Client {client_id}: {total_c} samples (Target: {int(total_size * volume_ratios[i])}) | Malignant ratio: {100 * num_m / total_c:.1f}%")

        clients.append((X_c, y_c))

    print("-" * 50)
    return clients


# ============================================================
# CLIENT CLASS (MLP / FedProx)
# ============================================================
class FLClient:
    def __init__(self, client_id, X_local, y_local):
        self.client_id = client_id
        # store as numpy arrays for Keras
        self.X = X_local.values.astype(np.float32)
        self.y = y_local.values.astype(np.float32)
        self.n_samples = len(self.X)
        self.local_model = None
        print(f"[Client {self.client_id}] Ready with {self.n_samples} samples.")

    def train_local_model(self, global_weights_t):
        """
        Build FedProx model, set weights (build first), then local train.
        """
        model = create_compiled_fedprox_model(MU, global_weights_t)
        # Build the model by calling it once so layers/weights exist
        _ = model(self.X[:1])
        model.set_weights(global_weights_t)

        model.fit(self.X, self.y,
                  epochs=MLP_EPOCHS,
                  batch_size=MLP_BATCH_SIZE,
                  verbose=0)

        return model

    def get_model_update_with_dp(self, global_weights_t):
        local_model = self.train_local_model(global_weights_t)
        updated_weights = local_model.get_weights()

        if USE_DP:
            # DP placeholder (not implemented)
            pass

        return updated_weights, self.n_samples


# ============================================================
# SERVER CLASS (FedAvg / FedProx aggregation)
# ============================================================
class FLServer:
    def __init__(self, clients, X_test, y_test, initial_global_weights):
        self.clients = clients
        self.X_test = X_test.values.astype(np.float32)
        self.y_test = y_test.values.astype(np.float32)
        self.history = []  # <-- fixed
        # store a deep copy of initial weights
        self.global_weights = [w.copy() for w in initial_global_weights]

        # persistent eval model (mu=0 for evaluation)
        self.eval_model = create_compiled_fedprox_model(0.0, initial_global_weights)
        # Build eval model before setting weights
        _ = self.eval_model(self.X_test[:1])
        self.eval_model.set_weights(self.global_weights)

    def sample_clients(self):
        # full participation
        return self.clients

    def run_training_round(self, round_num):
        selected_clients = self.sample_clients()
        all_client_updates = []  # <-- fixed

        for client in selected_clients:
            updated_weights, n_k = client.get_model_update_with_dp(self.global_weights)
            all_client_updates.append((updated_weights, n_k))

        # aggregate
        self.global_weights = self._aggregate_mlp_fedavg(all_client_updates)

        # evaluate on test
        self.eval_model.set_weights(self.global_weights)
        y_pred = self._get_predictions(self.eval_model)

        acc = accuracy_score(self.y_test, y_pred)
        sensitivity = recall_score(self.y_test, y_pred, pos_label=1)
        specificity = recall_score(self.y_test, y_pred, pos_label=0)
        f1 = f1_score(self.y_test, y_pred, pos_label=1)

        self.history.append((round_num, acc, sensitivity, specificity, f1))
        return acc, sensitivity, specificity, f1

    def _aggregate_mlp_fedavg(self, client_updates):
        """
        Weighted FedAvg aggregation: returns list of arrays matching weight shapes.
        """
        total_samples = sum(n_k for w_k, n_k in client_updates)
        # number of layers / tensors:
        num_tensors = len(client_updates[0][0])
        # initialize zeros with same shapes/dtypes as first client's weights
        global_weights = [np.zeros_like(client_updates[0][0][i]) for i in range(num_tensors)]

        for w_k, n_k in client_updates:
            alpha_k = n_k / total_samples
            for i in range(num_tensors):
                global_weights[i] += w_k[i] * alpha_k

        return global_weights

    def _get_predictions(self, model):
        return (model.predict(self.X_test, verbose=0) > 0.5).astype(int).flatten()

    def run(self, n_rounds):
        print(f"\n=== TRAINING START (MU={MU:.4f}) ===")
        for r in range(1, n_rounds + 1):
            if r % 5 == 0 or r == 1 or r == n_rounds:
                print(f"\n--- Round {r}/{n_rounds} ---")
                acc, sens, spec, f1 = self.run_training_round(r)
                print(f"ACC={acc:.4f}  SEN={sens:.4f}  SPEC={spec:.4f}  F1={f1:.4f}")
            else:
                self.run_training_round(r)
        return self.history


# ============================================================
# EXECUTION (Comparative Experiment under Feature Shift)
# ============================================================
if __name__ == "__main__":
    # --- Data Preparation ---
    X_train_full, y_train_full, X_test, y_test = load_and_preprocess_wdbc()
    client_partitions = create_non_iid_partitions(X_train_full, y_train_full, NUM_CLIENTS)

    # --- 1. Generate Initial Global Weights Structure (Reusable) ---
    initial_global_weights_zero = get_initial_weights(MLP_INPUT_DIM)

    # --- 2. Experiment 1: FedAvg Baseline (MU = 0.0) ---
    MU = 0.0  # FedAvg
    clients_fedavg = [FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_partitions)]
    server_fedavg = FLServer(clients_fedavg, X_test, y_test, initial_global_weights_zero)
    history_fedavg = server_fedavg.run(N_ROUNDS)

    # --- 3. Experiment 2: FedProx (MU = 0.01) ---
    MU = 0.01
    clients_fedprox = [FLClient(i + 1, Xc, yc) for i, (Xc, yc) in enumerate(client_partitions)]
    server_fedprox = FLServer(clients_fedprox, X_test, y_test, initial_global_weights_zero)
    history_fedprox = server_fedprox.run(N_ROUNDS)

    # --- 4. Final Comparison ---
    print("\n=== FINAL RESULTS ===")
    fedavg_final = history_fedavg[-1]
    fedprox_final = history_fedprox[-1]
    print(f"FedAvg Final: {fedavg_final}")
    print(f"FedProx Final: {fedprox_final}")
```
